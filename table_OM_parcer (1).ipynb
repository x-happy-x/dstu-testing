```python
import os
import re
from docx import Document
from docx.enum.table import WD_TABLE_ALIGNMENT

container = []
codes = {'Уровень 1': 'Знать: ', 'Уровень 2': 'Уметь: ', 'Уровень 3': 'Иметь навыки: ', 'Уровень 4': ''}

last_comp, last_ind = None, None

file_path = "./test-table.docx"

pr_table = Document()
table = pr_table.add_table(rows=1, cols=1, style='Table Grid')

for i in json_string['data']['items']:
    row_cells = table.add_row().cells
    if i['contents'] != None:
        comp = i['compCode'] + ' ' + i['compDescription']
        ind = i['indiCode'] + ' ' + i['indiDescription']

        if last_comp != comp:
            last_comp = comp
            row_cells[0].text = comp
            row_cells[0].alignment = WD_TABLE_ALIGNMENT.CENTER
            row_cells[0].paragraphs[0].paragraph_format.alignment = WD_TABLE_ALIGNMENT.CENTER
            row_cells[0].paragraphs[0].runs[0].bold = True
            row_cells = table.add_row().cells
        if last_ind != ind:
            last_ind = ind
            row_cells[0].text = ind
            row_cells[0].paragraphs[0].runs[0].bold = True
            row_cells = table.add_row().cells
        lvl = i['contents']
        lvl = re.sub(r'[\r\n\u200b]', ' ', lvl)
        row_cells[0].paragraphs[0].add_run(codes[i['code']]).bold = True
        row_cells[0].paragraphs[0].add_run(lvl)

pr_table.save(file_path)
```

```python
from docx import Document
from docx.shared import Pt
from docx.enum.section import WD_ORIENT
from docx.enum.section import WD_SECTION
from docx.shared import Cm

file_path = "./praktik-table.docx"

codes = {'Уровень 1': 'Запоминание', 'Уровень 2': 'Понимание ', 'Уровень 3': 'Применение ', 'Уровень 4': ''}

pr_table = Document()

current_section = pr_table.sections[-1]
new_width, new_height = current_section.page_height, current_section.page_width
new_section = pr_table.add_section(WD_SECTION.NEW_PAGE)
new_section.orientation = WD_ORIENT.LANDSCAPE
new_section.page_width = new_width
new_section.page_height = new_height

new_section.top_margin = Cm(2)
new_section.bottom_margin = Cm(1)
new_section.left_margin = Cm(1)
new_section.right_margin = Cm(1)

table = pr_table.add_table(rows=1, cols=7, style='Table Grid')
k = 0
for i in json_string['data']['items']:
    row_cells = table.add_row().cells
    if i['contents'] != None:
        comp = i['compCode'] + ' ' + i['compDescription']
        ind = i['indiCode'] + ' ' + i['indiDescription']

        lvl = i['contents']
        lvl = re.sub(r'[\r\n\u200b]', ' ', lvl)
        row_cells[1].paragraphs[0].add_run(codes[i['code']]).bold = True

        row_cells[2].paragraphs[0].add_run(lvl)

        row_cells[3].paragraphs[0].add_run('Контактная работа с преподавателем, самостоятельная работа')
        row_cells[3].paragraphs[0].paragraph_format.alignment = WD_TABLE_ALIGNMENT.CENTER

        row_cells[4].paragraphs[0].add_run('Разделы 1-3')
        row_cells[4].paragraphs[0].paragraph_format.alignment = WD_TABLE_ALIGNMENT.CENTER

        row_cells[5].paragraphs[0].add_run(
            'Отчет по практике, индивидуальное задание на практику, защита отчета по практике')
        row_cells[5].paragraphs[0].paragraph_format.alignment = WD_TABLE_ALIGNMENT.CENTER

        row_cells[6].paragraphs[0].add_run(
            'выполнение индивидуального задания на практику, выполнение отчета в назначенный срок, ответы на вопросы преподавателя по отчету по практике')
        row_cells[6].paragraphs[0].paragraph_format.alignment = WD_TABLE_ALIGNMENT.CENTER
    k += 1
    if k == 5:
        break

pr_table.save(file_path)
```
